```python
# encoding

from sklearn.preprocessing import LabelEncoder 

le=LabelEncoder()

df['Education_Level']=le.fit_transform(df['Education_Level'])
df['Attrition']=le.fit_transform(df['Attrition'])
df['Department']=le.fit_transform(df['Department'])
df
```

```python
# handle missing values 

# replace Numerical columns → mean

num_cols=["Performance_Score"]
df[num_cols]=df[num_cols].fillna(df[num_cols].mean())

# replace Categorical columns → mode

cat_cols=["Education_Level","Department"]
for col in cat_cols:
    df[col]=df[col].fillna(df[col].mode()[0]) # mode returns index therefore .mode()[0] means replace with the value at index 0 
```
