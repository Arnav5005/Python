```python
# dropping columns

df=df.drop(['employee_id'], axis=1)
df.info()
```

```python
df["education_level"].unique()
df["job_role"].unique()
```

```python
# encoding

# ≤ 10–15 unique values → One-hot is safe

# Very high (like 50+) → Consider frequency/target encoding


for col in ['gender','education_level','job_role','company_type','remote_work','deadline_frequency','physical_activity','health_issues']:
    print(col, df[col].nunique())

# from sklearn.preprocessing import LabelEncoder
# le=LabelEncoder()
# df['gender']=le.fit_transform(df['gender'])

# one-hot encoding

df = pd.get_dummies(
    df, 
    columns=['gender','education_level','job_role','company_type','remote_work','deadline_frequency','physical_activity','health_issues'], 
    drop_first=True
)

# Convert all boolean columns to 0/1 which are encoded

bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
# Heatmap vs TARGET ONLY

corr = df.corr()
target = 'burnout_score'

corr_target = corr[[target]].sort_values(
    by=target, ascending=False
)

plt.figure(figsize=(6, 10))
sns.heatmap(
    corr_target,
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title("Correlation with Burnout Score")
plt.tight_layout()
plt.show()
```

```python
# The 3-step rule (memorize this 🧠)
# ✅ Step 1: Build a simple baseline

# Regression → Linear Regression

# Classification → Logistic Regression

# Why?

# Verifies your pipeline works

# Gives a reference score

# Fast, interpretable

# ✅ Step 2: Build a strong baseline

# Random Forest

# Why?

# Captures non-linearity

# Handles messy real data

# Very strong out-of-the-box

# ✅ Step 3: Compare

# If RF ≫ Linear → data is non-linear

# If RF ≈ Linear → data is mostly linear

# If both bad → feature problem, not model
```

```python
from sklearn.model_selection import train_test_split

x=df.drop("burnout_score",axis=1)
y=df["burnout_score"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train,y_train)

y_pred=model.predict(x_test)
```
