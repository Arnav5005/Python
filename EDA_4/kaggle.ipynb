```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train_Kaggle_Comp.csv")
df
```

```python
# dropping columns

df=df.drop(['employee_id'], axis=1)
df.info()

# some people have experience more than their age so they need to be dropped

df=df.drop(df[df["age"]<df["experience_years"]].index)
```

```python
df["education_level"].unique()
df["job_role"].unique()
```

```python
for col in ['gender','education_level','job_role','company_type','remote_work','deadline_frequency','physical_activity','health_issues']:
    print(col, df[col].nunique())
```

```python
# one-hot encoding

df = pd.get_dummies(
    df, 
    columns=['gender','education_level','job_role','company_type','remote_work','deadline_frequency','physical_activity','health_issues'], 
    drop_first=True
)

bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
from sklearn.model_selection import train_test_split

x=df.drop("burnout_score",axis=1)
y=df["burnout_score"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train,y_train)

y_pred=model.predict(x_test)
```

```python
test=pd.read_csv("test_Kaggle_Comp.csv")
test=test.drop(["employee_id"],axis=1)

# encoding on test.csv

test=pd.get_dummies(
    test,
    columns=['gender','education_level','job_role','company_type','remote_work','deadline_frequency','physical_activity','health_issues'], 
    drop_first=True
)

bool_cols=test.select_dtypes(include='bool').columns
test[bool_cols]=test[bool_cols].astype(int)
```

```python
from sklearn.linear_model import LinearRegression

final_model=LinearRegression()
final_model.fit(x,y) 
```

```python
test_pred=final_model.predict(test)
```

```python
submission=pd.read_csv("sample_submission_Kaggle_Comp.csv")
submission["burnout_score"]=test_pred
submission.to_csv("submission.csv",index=False)
```
