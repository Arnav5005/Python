```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### encoding is the process of converting data, especially categorical data, into a numerical format that machine learning algorithms can understand and process

#### Label Encoding (assigns integers to categories)
#### Ordinal Encoding (assigns integers with a meaningful order)
#### One-Hot Encoding (creates a new binary column for each category)
#### Binary Encoding (encodes categories into binary values)
#### Other common types include Target Encoding (uses the target variable) and Frequency Encoding (uses the count of each category)
